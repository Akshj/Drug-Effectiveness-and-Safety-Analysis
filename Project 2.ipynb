```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv('drugs_side_effects_drugs_com.csv')
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
#Fill with meaningful labels
text_cols = ["side_effects", "generic_name", "drug_classes","brand_names", "pregnancy_category", "related_drugs"]
for col in text_cols:
             df[col] = df[col].fillna("Not Available")
df.isnull().sum()
```

```python
df.describe()
```

```python
#Numerical columns (rating, no_of_reviews):Since these are user-generated:


df["rating"] = df["rating"].fillna(df["rating"].median())
df["no_of_reviews"] = df["no_of_reviews"].fillna(0)              #Median avoids bias,0 reviews = logically correct
#Binary / category column (rx_otc)

df["rx_otc"] = df["rx_otc"].fillna("Unknown")

df.isnull().sum()

```

```python
#Handle alcohol column (Binary feature),1 → interacts with alcohol;0 → no info / no interaction
df['alcohol'] = df['alcohol'].fillna(0)
df['alcohol'] = df['alcohol'].replace({'X': 1})
```

Exploratory Data Analysis

2.Distribution Of Drug_ratings

```python
plt.figure(figsize=(8,5))
sns.histplot(df['rating'], bins=10, kde=True)
plt.title("Distribution of Drug Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

6.Side Effect Analysis

```python
df['side_effects'].head()
```

```python
#Split & Normalize Side Effects
import re

def split_side_effects(text):
    return [x.strip().lower() for x in re.split('[,;]', text)]

side_effects_series = (df['side_effects'].dropna().apply(split_side_effects).explode())
side_effects_series.value_counts().head(10)
```

6.1 Side Effects Vs Average Rating 

```python
top_side_effects = side_effects_series.value_counts().head(5).index

df['has_common_side_effect'] = df['side_effects'].str.lower().apply(
    lambda x: any(se in x for se in top_side_effects)
)

df.groupby('has_common_side_effect')['rating'].mean()
```

Drugs with common side effects tend to have slightly lower average ratings

Side effects impact user satisfaction

7.Drug Classes Analysis


```python
df['drug_classes'].head()
#Split Drug_classes
def split_drug_classes(text):
    return [x.strip() for x in text.split(',')]

drug_class_series = ( df['drug_classes'].dropna().apply(split_drug_classes).explode())
drug_class_series.value_counts().head(10)
```

Topical acne agents
Antihistamines
Upper respiratory combinations
Steroids
these drug classes, Indicates high-demand therapeutic areas

7.2 Drug Classes Vs Average rating

```python

drug_class_rating = (
    df.assign(drug_class=df['drug_classes'].str.split(','))
      .explode('drug_class')
      .groupby('drug_class')['rating']
      .mean()
      .sort_values(ascending=False)
)

drug_class_rating.head(10)
```
